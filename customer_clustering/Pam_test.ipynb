```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib as jlb

from sklearn_extra.cluster import KMedoids
from collections import Counter

#------------------------------------------------------------------------------
# Đọc dữ liệu
#------------------------------------------------------------------------------
data = pd.read_csv(folder + '/Data/Shopping.csv')
print(data.head())

#------------------------------------------------------------------------------
# Chuyển Genre sang số
#------------------------------------------------------------------------------
data = pd.get_dummies(data, columns=['Genre'], dtype=int, drop_first=True)
print(data.head())

#------------------------------------------------------------------------------
# Xây dựng mô hình PAM (K-Medoids)
#------------------------------------------------------------------------------
k = 5

model = KMedoids(n_clusters=k, random_state=42)
model.fit(data)

#------------------------------------------------------------------------------
# Lưu mô hình
#------------------------------------------------------------------------------
jlb.dump(model, folder + '/Output/PAM.mdl')

#------------------------------------------------------------------------------
# Load mô hình
#------------------------------------------------------------------------------
model = jlb.load(folder + '/Output/PAM.mdl')

#------------------------------------------------------------------------------
# Medoids
#------------------------------------------------------------------------------
print(f'*** {k} medoids:')
print(model.cluster_centers_)

#------------------------------------------------------------------------------
# Cluster labels
#------------------------------------------------------------------------------
data['cluster'] = model.labels_

print(data.head(10))

print('\nKích thước clusters:', Counter(model.labels_))
print('\n2 clusters lớn nhất:', Counter(model.labels_).most_common(2))

#------------------------------------------------------------------------------
# Visualization
#------------------------------------------------------------------------------
plt.figure(figsize=(8,8))

medoids = model.cluster_centers_

plt.scatter(medoids[:,0], medoids[:,1], marker='s', s=80, color='red')

plt.scatter(data['Age'], data['Annual Income'], c=data['cluster'])

plt.xlabel("Age")
plt.ylabel("Income")
plt.title("PAM Customer Segmentation", color='blue')

plt.show()
```
